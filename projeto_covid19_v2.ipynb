#**Panorama atual do Covid-19 no Mundo e no Brasil**

Os coronavírus são uma grande família de vírus comuns em muitas espécies diferentes de animais. Raramente, os coronavírus que infectam animais podem infectar pessoas. Recentemente, em dezembro de 2019, houve transmissão de um novo coronavírus (SARS-CoV-2), o qual foi identificado em Wuhan, na China, e causou a COVID-19, sendo em seguida disseminada e transmitida pessoa a pessoa.

<img alt="Cidade de Buenos Aires" width="50%" src='https://www.sbac.org.br/wp-content/uploads/2020/04/CONHECENDO-O-INIMIGO.png'>

A COVID-19 é uma doença causada pelo coronavírus, denominada SARS-CoV-2, que apresenta um espectro clínico variando de infecções assintomáticas a quadros graves. De acordo com a Organização Mundial da Saúde - OMS, a maioria (cerca de 80%) dos paciente com a doença podem ser assintomáticos ou ologossintomáticos (poucos sintomas), e aproximadamente 20% dos casos detectados requer atendimento hospitalar por apresentarem dificuldade respiratória, dos quais aproximadamente 5% podem necessitar de suporte ventilatório.

Os sintomas mais comuns em pessoas que contraem o COVID-19 são:

*   Tosse
*   Febre
*   Coriza
*   Dor de garganta
*   Dificuldade para respirar
*   Perda de olfato
*   Alteração do paladar
*   Disturbios gastrintetinais
*   Cansaço
*   Diminuição do apetite
*   Dispnéia

A transmissão do vírus ocorre:

*   Toque do aperto de mão contaminadas
*   Gotículas de Saliva
*   Espirro
*   Tosse
*   Catarro
*   Objetos ou superfícies contaminadas, como celulares, mesas, talheres, brinquedos, teclados de computador, etc.

Após algum tempo do aparecimento da doença, mais precisamente em 11 de março de 2020, a COVID-19 foi caracterizada pela OMS como uma pandemia. O termo se refere à distribuição geográfica de uma doenção e não à sua gravidade. A designação reconhece que, no momento, existem surtos do vírus em vários países e regiões do mundo.

Com o objetivo de elevar a consciência situacional do COVID-19 no mundo e no Brasil, iremos realizar uma analise sobre os dados públicos da doença.

Por se tratar de uma doença nova, a coleta de dados relacionados à doença ainda não possui uma metodologia clara e bem definida. Isso gera alguns gargalos no que diz respeito a informações referentes à doença.

*   `weekly_hosp_admissions` - número de pacientes recém-admitidos em hospitais em uma determinada semana
*   `weekly_hosp_admissions_per_million` - número de pacientes recém-admitidos em hospitais em uma determinada semana por milhão de habitantes
*   `total_tests` - número total de testes realizados

*   `tests_units` - unidade de medida utilizada por cada localidade para relatar os dados sobre testagem
*   `total_vaccinations` - número total de doses de vacinas administradas
*   `people_vaccinated` - número total de pessoas que receberam pelo menos uma dose da vacina
*   `people_fully_vaccinated` - número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação

##**Visualização dos Dados**

Após fazer uma breve exploração do nosso *dataset* chegou o momento de verificarmos alguns desses dados através dos *plots*. Isso permitirá que façamos algumas analises sobre a situação atual da pandemia causada pelo COVID-19.

###**TOP 5 em número de casos, mortes, novos casos e novas mortes**

Para iniciar o processo de visualização dos dados verificaremos quais são os cinco países que se encontram no topo da lista de casos, mortes, novos casos e novas mortes por COVID-19, no mundo. 

Vale ressalvar, nesse momento, que os dados utilizados para analise foram atualizados no dia 01 de Fevereiro de 2021, ou seja, todas as analises feitas aqui valem apenas para até essa data.

Vejamos a seguir:

```
#cinco países com maior número total de casos no dia 31 de janeiro de 2021
top5_total_cases = df.loc[df.date == '2021-01-31', ['location', 'total_cases']].sort_values(by='total_cases', ascending=False)[1:6]
```

```
#cinco países com maior número total de mortes no dia 31 de janeiro de 2021
top5_total_deaths = df.loc[df.date == '2021-01-31', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[1:6]
```

```
#cinco países com maior número de novos casos no dia 31 de janeiro de 2021
top5_new_cases = df.loc[df.date == '2021-01-31', ['location', 'new_cases']].sort_values(by='new_cases', ascending=False)[1:6]
```

```
#cinco países com maior número de novas mortes no dia 31 de janeiro de 2021
top5_new_deaths = df.loc[df.date == '2021-01-31', ['location', 'new_deaths']].sort_values(by='new_deaths', ascending=False)[1:6]
top5_new_deaths.head()
```

```
#Metodologia orientada a objetos para criação dos gráficos referentes ao número total de casos e mortes e ao número de novos casos e mortes

#criação da figure e dos axes
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,8))

#criação do gráfico do número total de casos no mundo
ax[0][0].bar('location', 'total_cases', data=top5_total_cases, color='darkred', label='Número Total de Casos')
ax[0][0].set_title('Top 5 - Países com maior número de casos de Covid-19', size=18)
ax[0][0].set_xlabel('Países', size=12)
ax[0][0].set_ylabel('Quantidade', size=12)
ax[0][0].legend()
#retirando a notação científica
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

#criação do gráfico do número de novos casos 
ax[0][1].bar('location', 'new_cases', data=top5_new_cases, color='red', label='Número de Novos Casos')
ax[0][1].set_title('Top 5 - Países com maior número de novos casos de Covid-19', size=18)
ax[0][1].set_xlabel('Países', size=12)
ax[0][1].set_ylabel('Quantidade', size=12)
ax[0][1].legend()

#criação do gráfico do número total de mortes
ax[1][0].bar('location', 'total_deaths', data=top5_total_deaths, color='orangered', label='Número Total de Mortes')
ax[1][0].set_title('Top 5 - Países com maior número de mortes por Covid-19', size=18)
ax[1][0].set_xlabel('Países', size=12)
ax[1][0].set_ylabel('Quantidade', size=12)
ax[1][0].legend()

#criação do gráfico do número de novas mortes
ax[1][1].bar('location', 'new_deaths', data=top5_new_deaths, color='darkorange', label='Número de Novas Mortes')
ax[1][1].set_title('Top 5 - Países com maior número de novas mortes por Covid-19', size=18)
ax[1][1].set_xlabel('Países', size=12)
ax[1][1].set_ylabel('Quantidade', size=12)
ax[1][1].legend()

#ajustando a saída dos gráficos
plt.tight_layout()
```

A partir da observação dos gráficos, com dados atualizados no dia 31 de Janeiro de 2021, podemos conncluir que:

*  Os Estados Unidos aparecem em primeiro em todos os quatro levantamentos realizados (número total de casos e mortes e número de novos casos e mortes);
*  O Brasil e o Reino Unidos aparecem entre os cinco maiores em todos os quatro levantamentos realizados;
*  A Rússia, a Índia, o Méixo e a França também estão na lista de países com destaque negativo na pandemia do Covid-19, aparecem entre os cinco primeiros paíes em dois ou três dos levantamentos realizados.




###**Evolução do Número de Casos e Mortes**

A partir de agora vejamos a evolução mundial no número total de casos e mortes por COVID-19. 

```
#fazendo uma cópia do dataframe
df_clean = df.copy()

#colocando a coluna date como nosso index
df_clean.set_index(df_clean.date, inplace=True)

#criando uma variável para o número total de casos no mundo
nun_cases = df_clean.loc[df_clean.location == 'World', 'total_cases']

#criando uma variável para o número total de mortes no mundo
num_deaths = df_clean.loc[df_clean.location == 'World', 'total_deaths']
```

```
#Metodologia de Abordagem Interface Pyplot

#criando um objeto do tipo Figure
plt.figure()

#construindo o gráfico com do número total de casos no Mundo
plt.subplot(221)
nun_cases.plot(x='date', y='num_cases', linewidth=2.25, color='orange', figsize=(22,8))
plt.title('Evolução dos casos de Covid-19 no Mundo', size=18)
plt.xlabel('Meses/Ano', size=14)
plt.ylabel('Quantidade', size=14)
plt.ylim(0)
plt.grid(True)
#modificando a notação científica
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

#construindo o gráfico com o número total de mortes no Mundo
plt.subplot(223)
num_deaths.plot(x='date', y='nun_deaths', linewidth=2.25, color='red', figsize=(22,8))
plt.title('Evolução das mortes por Covid-19 no Mundo', size=18)
plt.xlabel('Meses/Ano', size=14)
plt.ylabel('Quantidade', size=14)
plt.ylim(0)
plt.grid(True)
#modificando a notação científica
plt.gcf().axes[1].yaxis.get_major_formatter().set_scientific(False)

#ajustando a saíde
plt.tight_layout()

```

A observação e análise dos gráficos, nos ajudar a concluir que:

*  Após mais de um ano de início da Pandemia do Covid-19, a situação continua sendo bastante delicada;
*  Atingimos um número acima de 100 milhões de casos confirmados da doença;
*  Além disso, o número de mortes provocadas pela doenção ultrapassou a marca de 2 milhões de pessoas com destaque para os Estados Unidos e o Brasil, que eram os únicos países que tinham ultrapassado a marca de 200 mil mortes provocadas pela doença.

###**Evolução de Novos Casos e Novas Mortes**

A partir de agora veremos como anda a evolução do número de novos casos e de novas mortes por COVID-19.

```
#criando a variável de novos casos
new_cases = df_clean.loc[df_clean.location == 'World', 'new_cases']

#calculando a média móvel de sete dias para os novos casos
new_cases_media = new_cases.rolling(window=7).mean()
```

```
#criando a variável de novas mortes
new_deaths = df_clean.loc[df_clean.location == 'World', 'new_deaths']

#calculando a média móvel de sete dias para as novas mortes
new_deaths_media = new_deaths.rolling(window=7).mean()
```

```
#Abordagem Orientada a Objetos

#criação dos objetos Figure e Axes
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12,6))

#criando o gráfico de número de novos casos e média móvel de sete dias
ax[0].bar(new_cases.index, new_cases, color='orange', label='Nº de Novos Casos')
ax[0].plot(new_cases_media.index, new_cases_media, color='red', label='Média Móvel 7 dias')
ax[0].set_title('Número de Novos Casos de Covid-19', size=16)
ax[0].set_xlabel('Ano-Mês', size=12)
ax[0].set_ylabel('Quantidade', size=12)
ax[0].set_ylim(0)
ax[0].grid(True)
ax[0].legend()

#criando o gráfico de número de novas mortes e a média móvel de sete dias
ax[1].bar(new_deaths.index, new_deaths, color='orange', label='Nº de Novas Mortes')
ax[1].plot(new_deaths_media.index, new_deaths_media, color='red', label='Média Móvel 7 dias')
ax[1].set_title('Número de Novas Mortes por Covid-19', size=16)
ax[1].set_xlabel('Ano-Mês', size=12)
ax[1].set_ylabel('Quantidade', size=12)
ax[1].set_ylim(0)
ax[1].grid(True)
ax[1].legend()

#configurando a saídas dos gráficos
plt.tight_layout()
```

**Evolução do Processo de Vacinação no Mundo**

A partir de agora vamos analisar como se encontra o processo de vacinação a nível mundial.


```
#criação da variável de total de vacinas
total_vaccinations = df_clean.loc[df_clean.location == 'World', 'total_vaccinations']

#criação da variável do número de pessoas vacinadas
people_vaccinated = df_clean.loc[df_clean.location == 'World', 'people_vaccinated']

#criação da variável do número de novos vacinados
new_vaccinations = df_clean.loc[df_clean.location == 'World', 'new_vaccinations']

#media móvel de sete dias para o número de novos vacinados
new_vaccination_media = new_vaccinations.rolling(window=7).mean()
```

```
#criando os objetos do tipo Figure e Axes
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(12,8))

#gráfico do número total de vacinas
ax[0].plot(total_vaccinations.index, total_vaccinations, color='red', linewidth=1.5, label='Número total de vacinas')
ax[0].set_title('Número total de vacinas administradas', size=16)
ax[0].set_xlabel('Ano-Mês-Dia', size=12)
ax[0].set_ylabel('Quantidade', size=12)
ax[0].set_ylim(0)
ax[0].grid(True)
ax[0].legend()
#modificando a notação científica
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

#gráfico do número total de pessoas vacinadas
ax[1].plot(people_vaccinated.index, people_vaccinated, color='orange', linewidth=1.5, label='Número de pessoas vacinadas')
ax[1].set_title('Número total de pessoas vacinadas', size=16)
ax[1].set_xlabel('Ano-Mês-Dia', size=12)
ax[1].set_ylabel('Quantidade', size=12)
ax[1].set_ylim(0)
ax[1].grid(True)
ax[1].legend()
#retirando a notação científica
plt.gcf().axes[1].yaxis.get_major_formatter().set_scientific(False)

#gráfico do número de novos vacinados
ax[2].bar(new_vaccinations.index, new_vaccinations, color='yellow', label='Número de novos vacinados')
ax[2].plot(new_vaccination_media.index, new_vaccination_media, color='darkred', linewidth=1.5, label='Média Móvel 7 dias')
ax[2].set_title('Número de novos vacinados', size=16)
ax[2].set_xlabel('Ano-Mês-Dia', size=12)
ax[2].set_ylabel('Quantidade', size=12)
ax[2].set_ylim(0)
ax[2].grid(True)
ax[2].legend(loc=2)
#retirando a notação científica
plt.gcf().axes[2].yaxis.get_major_formatter().set_scientific(False)

#configurando a saída dos gráficos
plt.tight_layout()
```

A partir da observação e análise dos gráfico acima, podemos concluir que:

*   No final de Janeiro de 2021 estavamos atingindo a marca de 100 milhões de vacinas administradas;

A última vizualização sobre o panorama mundial da Covid-19 vamos utilizar um gráfico de correção `scatterplot` para observar se existe alguma correção entre o número total de mortes com o Produto Interno Bruto por pessoa.

Vejamos a seguir:

```
#criando a variável do número total de mortes e o PIB per capita
gdp_total_deaths = df.loc[df.date == '2021-01-31', ['location', 'total_deaths', 'gdp_per_capita']].sort_values(by='total_deaths', ascending=False)

#removendo os valores nulos
my_data = gdp_total_deaths.dropna(how='any', subset=['total_deaths', 'gdp_per_capita'])

#removendo a linha mundo
my_data.drop(62456, axis=0, inplace=True)
my_data.head()
```

```
#Metodologia Orientada a objetos

#criando os objetos Figure e Axes
fig, ax = plt.subplots(figsize=(10,6))

#plotando o gráfico scatterplot
ax.scatter(x='gdp_per_capita', y='total_deaths', data=my_data, s=7.5, color='red')
ax.set_title('Número de Óbitos x PIB per Capita', size=18)
ax.set_xlabel('PIB per Capita', size=12)
ax.set_ylabel('Número de Óbitos', size=12)
ax.set_ylim(0, 500000, 50000)
ax.set_xlim(0, 130000)
ax.grid(True)

#ajustando a saída do gráfico
plt.tight_layout()
```

###**Número de Novos Casos e Óbitos no Brasil**

Nesse tópico queremos apresentar a tendência de novos casos e óbitos no Brasil. Vejamos a seguir:

```
#criando a variável para novos casos
new_cases_brazil = df_brazil.loc[df_brazil.location == 'Brazil', ['date', 'new_cases']]

#calculando a média móvel de sete dias para o número de novos casos
new_cases_brazil_media = new_cases_brazil.rolling(window=7).mean()
```

```
#criando a variável para novos óbitos
new_deaths_brazil = df_brazil.loc[df_brazil.location == 'Brazil', ['date', 'new_deaths']]

#calculando a média móvel de sete dias para o número de novos óbitos
new_deaths_brazil_media = new_deaths_brazil.rolling(window=7).mean()
```

```
#criando os gráficos para apresentar o número de novos casos e óbitos para o Brasil

#criação dos objetos Figure e Axes
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(16,8))

#gráfico do número de novos casos
ax[0].bar(new_cases_brazil.date, new_cases_brazil.new_cases, color='orange', label='Nº de novos casos')
ax[0].plot(new_cases_brazil.date, new_cases_brazil_media, linewidth=1.5, color='red', label='Média Móvel 7 dias')
ax[0].set_title('Número de Novos Casos de Covid-19, Brasil', size=16)
ax[0].set_xlabel('Ano-Mês', size=12)
ax[0].set_ylabel('Quantidade', size=12)
ax[0].legend()
ax[0].grid(True)

#gráfico para o número de novos óbitos
ax[1].bar(new_deaths_brazil.date, new_deaths_brazil.new_deaths, color='orange', label='Nº de novos óbitos')
ax[1].plot(new_deaths_brazil.date, new_deaths_brazil_media, linewidth=1.5, color='red', label='Média Móvel 7 dias')
ax[1].set_title('Número de Novos Óbitos por Covid-19, Brasil', size=16)
ax[1].set_xlabel('Ano-Mês', size=12)
ax[1].set_ylabel('Quantidade', size=12)
ax[1].legend(loc=2)
ax[1].grid(True)

#configurando saída dos gráficos
plt.tight_layout()
```

A partir da análise dos gráficos podemos concluir que:

*   o número de novos casos, nos últimos dois meses (dezembro e janeiro), superou, frequentemente, a casa dos 40 mil novos casos;
*   a média móvel de sete dias para o número de novos casos, em janeiro, ficou acima de 50 mil;
*   o número de novos óbitos, no mês de janeiro de 2021, superou, constantemente, a casa de 1000 novos óbitos;
